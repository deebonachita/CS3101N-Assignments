# 1.
Provide reasons why each of the following equations is, or is not, a linear equation:

1. $\cos{\theta}x+4y+z = \sqrt{3}$
   - **Linear:** No, it is not linear.
   - **Reason:** The presence of \( \cos{\theta} \) makes it nonlinear.

2. $\cos{x}4y + z = \sqrt{3}$
   - **Linear**: No, it is not linear.
   - **Reason**: The product of $\cos{x}$ and $4y$ makes it nonlinear.

3. $x^{-1} + 7y + z = \sin{(\frac{\pi}{2})}$
   - **Linear**: No, it is not linear.
   - **Reason**: The presence of $x^{-1}$ (inverse of $x$) makes it nonlinear.

4. $x^1 + y^1 + z^1 = 0$
   - **Linear**: Yes, it is linear.
   - **Reason**: Each term is raised to the power of 1, and there are no products or powers other than 1. It follows the standard form of a linear equation.
