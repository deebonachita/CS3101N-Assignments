1. Without the use of any python libraries or modules, develop a function that can perform matrix addition, given two numpy matrices.
2. Without the use of any python libraries or modules, develop a function that can multiply two numpy matrices.
3. The rule of distributivity states that given two matrices $A$ and $B$ and a scalar, $k$, then $k(A+B)=kA+kB$. Instead of writing a proof mathematically, develop two codes for $k(A+B)$ and $kA+kB$.
4. Without using a python library or modules develop a function that can extract the diagonal of a numpy matrix.
5. Without using a python library or modules develop a function that can find a trace of a numpy matrix.

Solution to #1

```python
def matrix_addition(matrix_a, matrix_b):
    result = []
    for i in range(len(matrix_a)):
        row = []
        for j in range(len(matrix_a[0])):
            row.append(matrix_a[i][j] + matrix_b[i][j])
        result.append(row)
    return result
```

Solution to #2

```python
def matrix_multiplication(matrix_a, matrix_b):
    result = []
    for i in range(len(matrix_a)):
        row = []
        for j in range(len(matrix_b[0])):
            # Dot product of the ith row of A and the jth column of B
            element = sum(matrix_a[i][k] * matrix_b[k][j] for k in range(len(matrix_a[0])))
            row.append(element)
        result.append(row)
    return result
```

Solution to #3

```python
def distributive_property(matrix_a, matrix_b, scalar):
    result_left = scalar * matrix_addition(matrix_a, matrix_b)
    result_right = matrix_addition(scalar * matrix_a, scalar * matrix_b)
    return result_left, result_right
```

Solution to #4

```python
def extract_diagonal(matrix):
    result = []
    for i in range(len(matrix)):
        result.append(matrix[i][i])
    return result
```

Solution to #5

```python
def matrix_trace(matrix):
    result = sum(matrix[i][i] for i in range(len(matrix)))
    return result
```

# 5.
Find the rank and nullity of the matrix:
\[
\begin{bmatrix}
2 & 4 & 0 & 0 & 1 \\
6 & 3 & 5 & 1 & 0 \\
1 & 0 & 2 & 2 & 5 \\
1 & 1 & 1 & 1 & 1 \\
\end{bmatrix}
\]

To find the rank, we perform row reduction to obtain the row-echelon form:
\[
\begin{bmatrix}
1 & 2 & 0 & 0 & \frac{1}{2} \\
0 & -9 & 5 & 1 & -3 \\
0 & -2 & 2 & 2 & \frac{9}{2} \\
0 & -1 & 1 & 1 & \frac{3}{2} \\
\end{bmatrix}
\]

Now, we find the rank by counting the number of nonzero rows in the row-echelon form. The rank is 4.

The nullity is given by the difference between the number of columns and the rank. The nullity is \(5 - 4 = 1\).

Therefore, the rank of the matrix is 4, and the nullity is 1.
